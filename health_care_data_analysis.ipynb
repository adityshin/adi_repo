Health care data analysis project:

This study aims to analyze a comprehensive healthcare dataset containing patient
information such as Name, Age, Gender, Blood Type, Medical Condition, Date of
Admission, Doctor, Hospital, Insurance Provider, Billing Amount, Room Number,
Admission Type, Discharge Date, Medication, and Test Results. The analysis focuses
on understanding patient demographics, treatment patterns, hospital operations,
insurance coverage, and patient outcomes. Insights from this analysis will contribute to
improving patient care,optimizing hospital resources, and enhancing healthcare delivery.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
github_url = ('https://raw.githubusercontent.com/ChigicherlaVamsi2711/health-care-data/main/health_care')
df = pd.read_csv(github_url)
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the healthcare dataset into a DataFrame

# Plotting age distribution of patients
plt.figure(figsize=(10, 6))
df['Age'].plot(kind='hist', bins=20)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.show()
```

```python
# Plotting gender distribution of patients
plt.figure(figsize=(8, 6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution of Patients')
plt.ylabel('')
plt.show()
```

```python
# Plotting blood type distribution of patients
plt.figure(figsize=(8, 6))
df['Blood Type'].value_counts().plot(kind='bar')
plt.title('Blood Type Distribution of Patients')
plt.xlabel('Blood Type')
plt.ylabel('Number of Patients')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Plotting distribution of medical conditions
plt.figure(figsize=(10, 6))
df['Medical Condition'].value_counts().plot(kind='bar')
plt.title('Distribution of Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plotting correlation between room type/number and billing amount
plt.figure(figsize=(10, 6))
df.groupby('Room Number')['Billing Amount'].mean().plot(kind='line', marker='o')
plt.title('Correlation Between Room Number and Billing Amount')
plt.xlabel('Room Number')
plt.ylabel('Average Billing Amount')
plt.grid(True)
plt.show()
```

```python
# Plotting commonly used insurance providers
plt.figure(figsize=(10, 6))
df['Insurance Provider'].value_counts().nlargest(10).plot(kind='bar')
plt.title('Commonly Used Insurance Providers')
plt.xlabel('Insurance Provider')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plotting medication distribution for specific medical conditions
plt.figure(figsize=(12, 8))
df.groupby('Medical Condition')['Medication'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.title('Medication Distribution for Specific Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Number of Medications')
plt.xticks(rotation=45)
plt.legend(title='Medication')
plt.grid(True)
plt.show()
```

```python
# Plotting distribution of patients by age and gender
plt.figure(figsize=(12, 8))
df.groupby(['Age', 'Gender']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Distribution of Patients by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```

```python
# Plotting room allocation based on medical condition and admission type
plt.figure(figsize=(12, 8))
df.groupby(['Medical Condition', 'Admission Type']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Room Allocation Based on Medical Condition and Admission Type')
plt.xlabel('Medical Condition')
plt.ylabel('Number of Rooms')
plt.xticks(rotation=45)
plt.legend(title='Admission Type')
plt.grid(True)
plt.show()
```

```python
# Plotting correlation between medical condition and billing amount
plt.figure(figsize=(10, 6))
df.groupby('Medical Condition')['Billing Amount'].mean().plot(kind='bar')
plt.title('Correlation Between Medical Condition and Billing Amount')
plt.xlabel('Medical Condition')
plt.ylabel('Average Billing Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plotting average billing amount by admission type and insurance coverage
plt.figure(figsize=(12, 8))
df.groupby(['Admission Type', 'Insurance Provider'])['Billing Amount'].mean().unstack().plot(kind='bar')
plt.title('Average Billing Amount by Admission Type and Insurance Coverage')
plt.xlabel('Admission Type')
plt.ylabel('Average Billing Amount')
plt.xticks(rotation=0)
plt.legend(title='Insurance Provider')
plt.grid(True)
plt.show()
```
